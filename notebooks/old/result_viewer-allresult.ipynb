```python
restable=allrestable.copy()
restable['segmentor_params']=restable['segmentor_params'].astype(str)
restable['featureExtractor_params']=restable['featureExtractor_params'].astype(str)
restable=restable.loc[restable['featureExtractor_params'].str.contains('Simple')]
# restable=restable.loc[restable['featureExtractor_params'].str.contains('Cook')]
# restable=restable.loc[restable.type=='classic']

# restable=restable.loc[restable.evalution=="KFoldEval"]
restable.strategy.loc[restable.strategy.str.contains("FastFinder")]='FastFinder'
restable.strategy.loc[restable.segmentor=="Probabilistic"]='SimpleStrategy'
restable=restable.loc[restable.strategy=='SimpleStrategy']#FastFinder
# restable=restable.loc[restable.strategy=='FastFinder']#FastFinder
# print(len(restable),len(allrestable))
# res2=restable.groupby(['run_date','dataset','featureExtractor','featureExtractor_params','strategy','segmentor','segmentor_params','type']).mean()
res2=restable.groupby(['run_date','dataset','segmentor','type']).mean()
# display(HTML(res2.to_html()))
#res2.groupby(['dataset','runname','type']).std()
from IPython.display import display, HTML
# 'segmentor_params'
import pandas as pd
import numpy as np
def calccell(item):
    if np.isnan(item['mean']) or np.isnan(item['std']):
        return 'nan'
    return ('%d±%d')%(item['mean'],item['std'])
res3=res2.groupby(['dataset','segmentor','type']).agg({'accuracy':['mean','std'],'f1':['mean','std','count']})*100#.round(2)
# print(res3.first())
# # print(res4.keys())
list={}
for k,v in res3.iterrows():
    key=k[0:len(k)-1]
#     print(key)
    if not(key in list):
        list[key]={}
#     typ=k[len(k)-1]
#     list[key]={typ:{'accuracy':calccell(v['accuracy']),'f1':calccell(v['f1'])}}
#     list[key][k[len(k)-1]]={'accuracy':calccell(v['accuracy']),'f1':calccell(v['f1'])}
    list[key][k[len(k)-1]+"_acc"]=calccell(v['accuracy'])
    list[key][k[len(k)-1]+"_f1"]=calccell(v['f1'])
    list[key]["count"]=v['f1']['count']/100

df2=pd.DataFrame(list).transpose()
display(HTML(df2.to_html()))
```
